```python
from ultralytics import YOLO
import json


def model_train():
    model = YOLO("yolov8n.pt")

    model.train(data="/Users/tianlezhu/Waymo_Obj_Tracking/yolo_dataset/data.yaml", epochs=10,
                device='cpu')  # train the model
    model.val(data="/Users/tianlezhu/Waymo_Obj_Tracking/yolo_dataset/data.yaml")

    # get results all from test set
    results = model("/Users/tianlezhu/Waymo_Obj_Tracking/data/testing/val2020")
    return results


results = model_train()
```

```python
import matplotlib.pyplot as plt

def display_results(results, num_images=2):
    # Plot the first 'num_images' results
    for img in results[:num_images]:  
        plt.figure()
        plt.imshow(img.permute(1, 2, 0))  
        plt.show()
        
display_results(results)
```

```python
def filter_results_by_camera(results, camera_number):
    filtered_results = []
    for result in results:
        # Extract the camera number from the file path
        current_camera_number = os.path.basename(result.path).split('_')[-1].split('.')[0]
        if current_camera_number == f'camera{camera_number}':
            filtered_results.append(result)
    return filtered_results

camera1_results = filter_results_by_camera(results, '1')
print(camera1_results)
```

```python
import cv2
import os

def save_results_as_images(results, output_dir='training_output'):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Process each result
    for i, result in enumerate(results):
        img = result.orig_img

        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = box
            cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

        # Save the image
        output_file = os.path.join(output_dir, f'result_{i}.jpg')
        cv2.imwrite(output_file, img)
        
save_results_as_images(camera1_results)
```

```python
import cv2
import numpy as np
import os
from sort.sort import Sort

def track_and_visualize_images(images_folder, results, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    tracker = Sort()

    
    image_files = sorted([f for f in os.listdir(images_folder) if f.endswith('.jpg') or f.endswith('.png')])
    for frame_index, image_file in enumerate(image_files):
        frame = cv2.imread(os.path.join(images_folder, image_file))
        if frame is None:
            continue

            
        current_detections = []
        if frame_index < len(results):
            for box in results[frame_index].boxes.xyxy:
                x1, y1, x2, y2 = box
                current_detections.append([x1, y1, x2, y2, 1.0])  
        current_detections = np.array(current_detections)

        
        track_bbs_ids = tracker.update(current_detections)
        
        for track in track_bbs_ids:
            x1, y1, x2, y2, track_id = map(int, track)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, str(track_id), (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

            
        output_path = os.path.join(output_folder, image_file)
        cv2.imwrite(output_path, frame)

        
images_folder = "/Users/tianlezhu/Waymo_Obj_Tracking/testing/camera1"
output_folder = "/Users/tianlezhu/Waymo_Obj_Tracking/output_images"
track_and_visualize_images(images_folder, camera1_results, output_folder)
```
